```python
import pandas as pd
```

```python
df = pd.read_csv('food-truck-data.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# remove the columns dayshours and NOISent
df = df.drop(['dayshours', 'NOISent'], axis=1)
```

```python
df.isnull().sum()
```

```python
# also remove block, lot, location
df = df.drop(['block', 'lot', 'Location'], axis=1)
df.head()
```

```python
# only keep applicant, facility type, status, food items, latitude, and longitude
df = df[['Applicant', 'FacilityType', 'Status', 'FoodItems', 'Latitude', 'Longitude']]
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# remove rows with missing values
df = df.dropna()
```

```python
df.shape
```

```python
# export the cleaned data
df.to_csv('food-truck-data-cleaned.csv', index=False)
```
